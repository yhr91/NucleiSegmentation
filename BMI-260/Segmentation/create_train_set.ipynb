## Script for creating training patches

- Divide patients into train and test
- Link patient images to binary masks
- Pull out 64x64 samples from these images using the binary masks

```python
from image_process import *
import matplotlib.pyplot as plt
import csv
%matplotlib inline
```

```python
data_dir = '/home/yr897021/Janowczyk/Data/'
norm_path = 'normalized_full/'
#filenames = read_filenames(data_dir + 'normalized/')
filenames = read_filenames(data_dir + norm_path)

# Extract context for each of these images
pix_per_image = 2133
patch_size = 64
orig_train_size = 96
train,test = [], []
```

```python
# Read in all images and extract patient ID's, create  a dict to link image to mask
nuc_images = {}; patients = []
norm_tag = 'NormRH'

train = read_list('train_patients')
test = read_list('test_patients')

for f in filenames:
    if '.tif' in f and norm_tag in f:
        f = f.strip(norm_tag)
        pat_id = f.split('_')[0]
        if pat_id in train:
            image_id = f.split('_original')[0]
            neg_path = data_dir + 'NegMasks_64/' + f + '_NegMask.csv'
            edge_path = data_dir + 'EdgeMasks/' + image_id + '_mask.png' + '_EdgeMask.csv'
            if pat_id not in patients:
                patients.append(pat_id)
            nuc_images[image_id] = {'image': cv2.imread(data_dir + norm_path + norm_tag + f),
                           'mask': cv2.imread(data_dir + image_id + '_mask.png'),
                           'neg_mask': csv2matrix(neg_path),
                           'edge_mask': csv2matrix(edge_path)}

# Split train and test patients
#train, test = train_test_split(patients, 0.7)
```

## Some helper functions

```python
def labelled_patch_matrix(patches, label):

    # Create a single list of all patches from all images-
    patch_list = np.array([np.array(p.values()) for p in patches if p is not None])

    # Flatten it into a one dimensional list of patches only
    patch_list_flat = np.array(lists_to_list(patch_list))
    X = np.array(patch_list_flat)

    # The labels should be straightforward
    y = np.multiply(np.ones(X.shape[0], dtype=np.int64), label)
    
    return X, y
```

## Create patches from nuclei regions

```python
# Create a list of nuclei pixels for each image
for item in nuc_images.itervalues():
    if item['mask'] is not None:
        item['nuclei'] = mask2pix(item['mask'][:,:,0]).values()

# Generate lists of patches from each image using pixel indices for nuclei
train_patches = []; test_patches = []
for key, val in nuc_images.iteritems():
    if val['mask'] is not None:
        if key.split('_')[0] in train:
            train_patches.append(context_extractor(val['image'], val['nuclei'][0], patch_size, pix_per_image))
        elif key.split('_')[0] in test:
            test_patches.append(context_extractor(val['image'], val['nuclei'][0], patch_size, pix_per_image))
        
X_nuc_train, y_nuc_train = labelled_patch_matrix(train_patches, 1)
X_nuc_test, y_nuc_test = labelled_patch_matrix(test_patches, 1)
```

## Create patches from non-nuclei regions

```python
# Just functionalizing the patching conditions for non nuclei regions
def non_nuclei_patch(item, patches):
    if item['edge_mask'] is 0 and item['neg_mask'] is not 0:
        patches.append(context_extractor(item['image'], item['normal_bkg'][0], patch_size, int(pix_per_image)))
    elif item['neg_mask'] is 0:
        return
    else:
        patches.append(context_extractor(item['image'], item['normal_bkg'][0], patch_size, int(0.75*pix_per_image)))
        patches.append(context_extractor(item['image'], item['normal_edg'][0], patch_size, int(0.25*pix_per_image)))
```

```python
# Create a list of non-nuclei pixels for each image
for item in nuc_images.itervalues():
    if item['mask'] is not None:
        if item['neg_mask'] is not 0:
            item['normal_bkg'] = mask2pix(item['neg_mask']).values()
        if item['edge_mask'] is not 0:
            item['normal_edg'] = mask2pix(item['edge_mask']).values()

# Generate lists of patches from each image using pixel indices for nuclei
train_patches = []; test_patches = []
for key, val in nuc_images.iteritems():
    if val['mask'] is not None:
        if key.split('_')[0] in train:
            non_nuclei_patch(val, train_patches)
        elif key.split('_')[0] in test:
            non_nuclei_patch(val, test_patches)
        
X_nor_train, y_nor_train = labelled_patch_matrix(train_patches, 0)
X_nor_test, y_nor_test = labelled_patch_matrix(test_patches, 0)
```

## Shuffle and save as an lmdb

```python
def save_lmdb(X,y,name):
    indices = range(X.shape[0])
    shuffle(indices)
    
    # Shuffle
    X_shuffled = X[indices]
    y_shuffled = y[indices]
    
    print X_shuffled.shape
    
    # -- Write labels
    lmdb_list = [str(i) + '.jpg ' + str(y_shuffled[i]) for i in range(X.shape[0])]
    write_lines(name + '_labels', lmdb_list)
    
    # --  Save LMDB 
    for i, img in enumerate(X_shuffled):
        cv2.imwrite('DB/' + name + '/' + str(i) +'.jpg', img)
        if (i%50000 == 0):
            print i
```

```python
X_full_train = np.concatenate((X_nuc_train, X_nor_train), axis=0)
y_full_train = np.concatenate((y_nuc_train, y_nor_train), axis=0)
save_lmdb(X_full_train, y_full_train,'NormRH_fix')
```

```python
X_full_test = np.concatenate((X_nuc_test, X_nor_test), axis=0)
y_full_test = np.concatenate((y_nuc_test, y_nor_test), axis=0)
save_lmdb(X_full_test, y_full_test,'DB_1_64_test_fix')
```
